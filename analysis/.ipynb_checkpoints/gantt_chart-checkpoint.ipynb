```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# ---------- CONFIGURATION ----------
# Style settings to match the user's example image, but with multiple colors
plt.rcParams.update({
    "font.family": "sans-serif",
    "font.sans-serif": "Arial",
    "figure.facecolor": "white",
    "axes.facecolor": "white",
    "axes.grid": True,
    "grid.color":"black",
    "grid.linestyle": ":",
    "grid.linewidth": 0.7,
    "xtick.color": "#495057",
    "ytick.color": "#212529",
    "axes.labelcolor": "#212529",
    "text.color": "#212529",
    "figure.dpi": 300
})

# Color palette for each Work Package
WP_COLORS = {
    "WP1": "#0d5aa2",  # Deep Blue
    "WP2": "#008450",  # Forest Green
    "WP3": "#e2711d",  # Burnt Orange
    "WP4": "#c72c41",  # Muted Red
}
WP_HEADER_BG_COLOR = "#F0F0F0"
WP_HEADER_TEXT_COLOR = "#212529"

# --- DATA (Original content with structured WPs for color assignment) ---
wps_structured_data = [
    {
        "label": "WP1 Semantic Parsing", "color": WP_COLORS["WP1"],
        "tasks": [
            {"label": "T1.1 Worry Prompt Task", "start": 1, "end": 8},
            {"label": "T1.2 Semantic Processing", "start": 5, "end": 13},
            {"label": "T1.3 MDP Bias Validation", "start": 10, "end": 18},
        ]
    },
    {
        "label": "WP2 Generalization Model", "color": WP_COLORS["WP2"],
        "tasks": [
            {"label": "T2.1 Model Development", "start": 12, "end": 24},
            {"label": "T2.2 Task Development", "start": 16, "end": 24},
            {"label": "T2.3 Empirical Testing and Analysis", "start": 24, "end": 30},
        ]
    },
    {
        "label": "WP3 Meta-Control Models", "color": WP_COLORS["WP3"],
        "tasks": [
            {"label": "T3.1 Initiation Failure Model Development", "start": 18, "end": 30},
            {"label": "T3.2 Initiation Failure Task Development", "start": 22, "end": 30},
            {"label": "T3.3 Termination Failure Model Development", "start": 18, "end": 30},
            {"label": "T3.4 Termination Failure Task Development", "start": 22, "end": 30},
            {"label": "T3.5 Empirical Testing and Analysis", "start": 30, "end": 36},
        ]
    },
    {
        "label": "WP4 Clinical Taxonomy", "color": WP_COLORS["WP4"],
        "tasks": [
            {"label": "T4.1 Piloting Parameter Stability", "start": 33, "end": 40},
            {"label": "T4.2 Longitudinal Study", "start": 40, "end": 44},
            {"label": "T4.3 Longitudinal Data Analysis", "start": 44, "end": 48},
        ]
    },
]

# Flatten the structured data into a list of rows for plotting
rows_data = []
for wp in wps_structured_data:
    rows_data.append({"label": wp['label'], "type": "wp_header"})
    for task in wp['tasks']:
        rows_data.append({
            "label": f"  {task['label']}",
            "type": "task",
            "start": task['start'],
            "end": task['end'],
            "color": wp['color']  # Assign the parent WP's color
        })

# MODIFICATION 1: Update phase data
# - Phase 1 starts at month 0
# - Phase 3 label is changed
phases_header = [
    {"label": "P1: World Model Extraction", "start_month": 0, "end_month": 18},
    {"label": "P2: Link Computations to Worry", "start_month": 18, "end_month": 36},
    {"label": "P3: Parameter Reliability", "start_month": 36, "end_month": 48},
]

MAX_MONTH = 48

# ---------- PLOTTING ----------
fig, ax = plt.subplots(figsize=(18, 10), layout="constrained")

# --- Draw Phase Headers (top-most section) ---
phase_header_height = 1.5
phase_header_y_start = len(rows_data) + 0.1 

# Draw the phase backgrounds and text first
for ph in phases_header:
    ax.add_patch(mpatches.Rectangle(
        (ph["start_month"], phase_header_y_start), 
        ph["end_month"] - ph["start_month"], 
        phase_header_height,
        facecolor='#e9ecef', 
        edgecolor='none',zorder=3
    ))

    # MODIFICATION 2: Adjust text position for Phase 3
    # Default text position (centered)
    text_x = (ph["start_month"] + ph["end_month"]) / 2
    text_ha = 'center'

    # Special handling to move Phase 3 text to the right
    if "Phase 3" in ph["label"]:
        text_x = ph["end_month"] - 0.5  # Position near the right edge with padding
        text_ha = 'right'              # Right-align the text

    ax.text(
        text_x, 
        phase_header_y_start + phase_header_height / 2, 
        ph["label"],
        ha=text_ha, # Use the determined horizontal alignment
        va='center', 
        fontsize=15, 
        weight='bold', 
        color='#212529',
        zorder=3
    )

# MODIFICATION 3: Outer border now starts at month 0
# Create an outer border for the entire phase section
overall_start = min(p['start_month'] for p in phases_header)
overall_end = max(p['end_month'] for p in phases_header)
ax.add_patch(mpatches.Rectangle(
    (overall_start, phase_header_y_start),
    overall_end - overall_start,
    phase_header_height,
    facecolor='none',
    edgecolor='#adb5bd',
    linewidth=1.0,zorder=3
))

# Explicitly draw vertical lines between phases
y_coords = [phase_header_y_start, phase_header_y_start + phase_header_height]
ax.plot([18, 18], y_coords, color='black', linewidth=1.0,zorder=3)
ax.plot([36, 36], y_coords, color='black', linewidth=1.0,zorder=3)


# --- Draw Gantt Bars ---
for i, row in enumerate(rows_data):
    if row['type'] == 'wp_header':
        ax.add_patch(mpatches.Rectangle(
            (0, i - 0.5), MAX_MONTH + 2, 1.0,
            facecolor=WP_HEADER_BG_COLOR, edgecolor='none', zorder=3
        ))
        ax.text(0.5, i, row['label'], ha='left', va='center',
                fontsize=20, weight='bold', color=WP_HEADER_TEXT_COLOR)
    elif row['type'] == 'task':
        start, end = row['start'], row['end']
        ax.barh(i, end - start, left=start, height=0.6,
                color=row['color'], edgecolor='none')

# --- Axis Formatting ---
ax.set_xlim(0, MAX_MONTH + 1)
ax.set_ylim(-0.5, len(rows_data) + 1.5)
ax.invert_yaxis()

# Y-axis labels
ax.set_yticks(range(len(rows_data)))
ax.set_yticklabels([row['label'] for row in rows_data], fontsize=20)
ax.tick_params(axis='y', length=0, pad=5)

# Hide the y-tick labels for WP headers to avoid repetition
counter=0
for i, row in enumerate(rows_data):
    if row['type'] == 'wp_header':
        ax.get_yticklabels()[i-counter].set_visible(False)
        counter+=1

# X-axis (single, clean axis at the bottom)
ax.set_xticks(np.arange(0, MAX_MONTH + 1, 6))
ax.set_xticklabels([f"M{i}" for i in np.arange(0, MAX_MONTH + 1, 6)], fontsize=15)
ax.set_xlabel("Project Months", fontsize=20, weight='bold', labelpad=10)
ax.tick_params(axis='x', length=0)

# Remove chart borders (spines)
ax.spines[['top', 'right', 'left', 'bottom']].set_visible(False)

# Add a title
fig.suptitle("Project Timeline and Work Package Breakdown", fontsize=22, weight='bold')
plt.savefig('GanttChart.png',dpi=600)
plt.show()
```
